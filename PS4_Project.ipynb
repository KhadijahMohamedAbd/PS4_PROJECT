# checking for outliers

```python
#Evaluating quartiles
ps4.quantile([0,0.25,0.5,0.75,1])
```

```python
ps4.boxplot()
```

```python
#InterQuartile Range
North_America_IQR=ps4['North America'].quantile(0.75)-ps4['North America'].quantile(0.25)
North_America_IQR
```

```python
#Europe
Europe_IQR=ps4['Europe'].quantile(0.75)-ps4['Europe'].quantile(0.25)
Europe_IQR
```

```python
#JAPAN
Japan_IQR=ps4['Japan'].quantile(0.75)-ps4['Japan'].quantile(0.25)
Japan_IQR
```

```python
#REST_OF_WORLD
Rest_World_IQR=ps4['Rest of World'].quantile(0.75)-ps4['Rest of World'].quantile(0.25)
Rest_World_IQR
```

```python
#GLOBAL
Global_IQR=ps4['Global'].quantile(0.75)-ps4['Global'].quantile(0.25)
Global_IQR
```

```python
#Outliers
America_lower=ps4['North America'].quantile(0.25)-(1.5*North_America_IQR)
America_lower
America_upper=ps4['North America'].quantile(0.75)+(1.5*North_America_IQR)
America_upper
```

```python
#outliers_for_north_america
America_Outliers=ps4[(ps4['North America']>America_upper) | (ps4['North America']<America_lower)]
America_Outliers
```

```python
#Europe_outliers
Europe_lower=ps4['Europe'].quantile(0.25)-(1.5*North_America_IQR)
Europe_lower
Europe_upper=ps4['Europe'].quantile(0.75)+(1.5*North_America_IQR)
Europe_upper
```

```python
#outliers_for_Europe
Europe_Outliers=ps4[(ps4['Europe']>Europe_upper) | (ps4['Europe']<Europe_lower)]
Europe_Outliers
```

```python
#Global_outliers
Global_lower=ps4['Global'].quantile(0.25)-(1.5*North_America_IQR)
Global_lower
Global_upper=ps4['Global'].quantile(0.75)+(1.5*North_America_IQR)
Global_upper
```

```python
#outliers_for_the_globe
Global_Outliers=ps4[(ps4['Global']>Global_upper) | (ps4['Global']<Global_lower)]
Global_Outliers
```

```python
#Japan_outliers
Japan_lower=ps4['Japan'].quantile(0.25)-(1.5*Japan_IQR)
Japan_lower
Japan_upper=ps4['Japan'].quantile(0.75)+(1.5*Japan_IQR)
Japan_upper
```

```python
#outliers_for_Japan
Japan_Outliers= ps4[(ps4['Japan']>Japan_upper) | (ps4['Japan']<Japan_lower)]
Japan_Outliers
```

```python
#Removing_japan_ Outliers
ps4=ps4[~(ps4['Japan']>Japan_upper)|(ps4['Japan']<Japan_lower)]
ps4
```

```python
#Removing Europe outliers
ps4=ps4[~(ps4['Europe']>Europe_upper) | (ps4['Europe']<Europe_lower)]
ps4
```

```python
#Removing_North_America_outliers
ps4=ps4[~(ps4['North America']>America_upper) | (ps4['North America']<America_lower)]
ps4
```

```python
#Removing Global outliers
ps4=ps4[~(ps4['Global']>Global_upper) | (ps4['Global']<Global_lower)]
ps4
```

# First Analyzing by Genre

```python
Genre_data=ps4.groupby('Genre').sum().sort_values('Global',ascending=False).reset_index()
Genre_data
```

```python
sns.barplot(x=Genre_data["Global"], y=Genre_data["Genre"], data=Genre_data, label="Global", color="b")
plt.title('PS4 BY GENRE GLOBALLY')
```

```python
Genre_North= ps4.groupby('Genre').sum().sort_values('North America',ascending=False).reset_index()
Genre_North
```

```python
sns.barplot(x=Genre_North["North America"], y=Genre_North["Genre"], data=Genre_North, color="b")
plt.title('PS4 BY GENRE North America')
```

```python
Genre_Europe= ps4.groupby('Genre').sum().sort_values('Europe',ascending=False).reset_index()
Genre_Europe
```

```python
sns.barplot(x=Genre_Europe["Europe"], y=Genre_Europe["Genre"], data=Genre_Europe, color="b")
plt.title('PS4 BY GENRE Europe Total')
```

```python
Genre_Japan= ps4.groupby('Genre').sum().sort_values('Japan',ascending=False).reset_index()
Genre_Japan
```

```python
sns.barplot(x=Genre_Japan["Japan"], y=Genre_Japan["Genre"], data=Genre_Japan, color="b")
plt.title('PS4 BY GENRE Japan Total')
```

**so, the top 5 Genres in North America, Japan, Europe and Globally are:**
    
    1-Racing
    2-Action
    3-shooter
    4-Action Adventure
    5-Role Playing
